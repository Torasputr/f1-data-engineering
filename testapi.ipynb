```python
import fastf1
import matplotlib.pyplot as plt

# fastf1.Cache.enable_cache("cache")  # folder cache lokal

session = fastf1.get_session(2024, "Monza", "Q")
session.load()

print(session.event["EventName"])
print(session.results[["Abbreviation", "TeamName", "Position"]].head())

laps_ver = session.laps.pick_driver("VER")
fastest_ver = laps_ver.pick_fastest()

car = fastest_ver.get_car_data().add_distance()

plt.figure(figsize=(10, 4))
plt.plot(car["Distance"], car["Speed"])
plt.title("VER Fastest Lap Speed Trace - Monza Q 2024")
plt.xlabel("Distance (m)")
plt.ylabel("Speed (km/h)")
plt.grid(True, alpha=0.3)
plt.show()
```
